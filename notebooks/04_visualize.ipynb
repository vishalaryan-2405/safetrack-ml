```python
import sys
sys.path.append('..')
import pandas as pd
import folium
from folium.plugins import HeatMap
import config as cfg

df = pd.read_csv('../data/scored_points.csv')
print("Points loaded:", len(df))
print("Anomalies:", df['anomaly'].sum())
print("Safe points:", (~df['anomaly']).sum())
```

```python
m = folium.Map(location=list(cfg.MAP_CENTER), zoom_start=13)

for _, row in df.iterrows():
    color  = 'red'   if row['anomaly'] else 'green'
    radius = 7       if row['anomaly'] else 4
    folium.CircleMarker(
        location=[row['lat'], row['lng']],
        radius=radius,
        color=color,
        fill=True,
        fill_opacity=0.7,
        popup=f"hour={int(row['hour'])}  anomaly={row['anomaly']}"
    ).add_to(m)

m
```

```python
for name, coords in cfg.LOCALITIES.items():
    folium.Marker(
        location=list(coords),
        popup=name.replace('_',' ').title(),
        tooltip=name.replace('_',' ').title(),
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)

m
```

```python
heat_data = [[row['lat'], row['lng']] for _, row in df.iterrows()]
HeatMap(heat_data, radius=15, blur=10, min_opacity=0.4).add_to(m)
m
```

```python
m.save('../maps/safety_map.html')
print("Map saved to maps/safety_map.html")
print("Open that file in Chrome to see the full interactive map!")
```
