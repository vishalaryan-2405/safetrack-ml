```python
import sys
sys.path.append('..')
import pandas as pd
import numpy as np
import config as cfg
np.random.seed(42)
print("Ready")
```

```python
def simulate_zone(center, n=100, spread=0.004, hours=range(8,20)):
    lats  = center[0] + np.random.normal(0, spread, n)
    lngs  = center[1] + np.random.normal(0, spread, n)
    hrs   = np.random.choice(list(hours), n)
    return pd.DataFrame({'lat': lats, 'lng': lngs, 'hour': hrs,
                         'zone': 'known'})
print("Function ready")
```

```python
frames = []
for name, coords in cfg.LOCALITIES.items():
    frames.append(simulate_zone(coords, n=80))

frames.append(simulate_zone(cfg.MY_HOME, n=120, hours=range(20,24)))

df_train = pd.concat(frames, ignore_index=True)
print(f"Training points: {len(df_train)}")
df_train.head()
```

```python
anomalies = pd.DataFrame({
    'lat' : [18.08, 18.05, 18.22, 18.19, 18.09],
    'lng' : [83.45, 83.35, 83.41, 83.38, 83.50],
    'hour': [2,     23,    3,     1,     22   ],
    'zone': ['anomaly'] * 5
})
df_all = pd.concat([df_train, anomalies], ignore_index=True)
print("Total points:", len(df_all))
anomalies
```

```python
df_train.to_csv('../data/training_gps.csv', index=False)
df_all.to_csv('../data/all_points.csv', index=False)
print("Saved! Check your data/ folder")
```
